# Exercice 3: Matrices, valeurs propres et vecteurs propres

## 1. Définition des matrices
Rappel : Une matrice carrée $A$ admet des **valeurs propres** $\lambda$ telles que :
$A v = \lambda v$ où $v$ est un vecteur propre.

```python
# ==========================
# 🧮 Exercice SymPy - Matrices
# ==========================

import sympy as sp
from IPython.display import display, Math

# ==========================
# 📌 Définition des matrices
# ==========================
# définir la matrice A
A = sp.Matrix([[2, 1], [1, 3]])

# définir la matrice B
B = sp.Matrix([[3, -1], [1, 2]])

# ==========================
# 📌 Affichage des matrices
# ==========================
# afficher A en LaTeX
display(Math(r"A = " + sp.latex(A)))

display(Math(r"B = " + sp.latex(B)))
```

## 2. Affichage des matrices

```python
# ==========================
# 📌 Affichage des matrices
# ==========================
# afficher A avec pprint (affichage propre)
print("Matrice A :")
sp.pprint(A)

# afficher B avec pprint (affichage propre)
print("\nMatrice B :")
sp.pprint(B)
```

## 3. Valeurs propres et vecteurs propres
Rappel :
- Les valeurs propres sont solutions de $\det(A - \lambda I)=0$
- Les vecteurs propres sont associés à chaque valeur propre.

```python
# ==========================
# 🧮 Valeurs propres et vecteurs propres (A)
# ==========================

import sympy as sp
from IPython.display import display, Math

# ==========================
# 📌 Définition de la matrice
# ==========================
A = sp.Matrix([[2, 1], [1, 3]])

# ==========================
# 📌 Calcul des valeurs propres
# ==========================
eig_A = A.eigenvects()

# ==========================
# 📌 Affichage des résultats
# ==========================
display(Math(r"\text{Valeurs propres et vecteurs propres de } A :"))

for val, mult, vects in eig_A:
    latex_val = sp.latex(val)
    latex_mult = sp.latex(mult)
    latex_vecs = ", ".join([sp.latex(v) for v in vects])

    display(Math(
        rf"\lambda = {latex_val},\quad "
        rf"\text{{multiplicité }} {latex_mult},\quad "
        rf"\text{{vecteurs propres : }} {latex_vecs}"
    ))
```
