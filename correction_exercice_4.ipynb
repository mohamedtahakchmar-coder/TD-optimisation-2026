## 🔹 Rappels mathématiques

### 1. Matrice symétrique
Une matrice $A$ est symétrique si :
$$ A^T = A $$

### 2. Déterminant
Le déterminant d'une matrice permet de savoir si elle est inversible.

### 3. Valeurs propres
Les valeurs propres $\lambda$ sont solutions de :
$$ \det(A - \lambda I) = 0 $$

### 4. Matrice définie positive
Une matrice est définie positive si :
- Toutes ses valeurs propres sont strictement positives
- Elle admet une décomposition de Cholesky

### 5. Décomposition de Cholesky
Une matrice définie positive peut s’écrire :
$$ A = LL^T $$
où $L$ est triangulaire inférieure.


## 🔹 1) Définir la matrice A

```python
# ==========================
# 🧮 Matrice 3x3
# ==========================

import sympy as sp

# ==========================
# 📌 Définition de la matrice
# ==========================
# définir la matrice A (3x3)
A = sp.Matrix([[4, 1, 1], [1, 3, 1], [1, 1, 2]])
```

## 🔹 2) Afficher la matrice

```python
print("Transposée de A :")
A_T = A.T
sp.pprint(A_T)
```

## 🔹 3) Calcul du déterminant

```python

# Calculer le déterminant de A  
det_A = A.det()
print("Déterminant de A :", det_A)
```

## 🔹 4) Calcul de la transposée

```python

A_T = A.T
print("Transposée de A :")
sp.pprint(A_T)
```

## 🔹 5) Valeurs propres

```python
# Calculer les valeurs propres de A  
eigenvalues = A.eigenvals()
print("Valeurs propres de A :")
sp.pprint(eigenvalues)
```

Si la décomposition de Cholesky échoue → la matrice n’est pas définie positive.
