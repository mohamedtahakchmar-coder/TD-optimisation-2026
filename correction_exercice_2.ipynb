# Exercice 2: Analyse de fonctions de deux variables

## 1. Définition des variables et des fonctions
Rappel : une fonction de deux variables $f(x,y)$ dépend simultanément de $x$ et $y$.

```python
import sympy as sp
from IPython.display import display, Math

# ==========================
# 📌 Variables symboliques
# ==========================
# définir les variables x et y comme réelles
x, y = sp.symbols('x y', real=True)

# ==========================
# 📌 Définition des fonctions
# ==========================
# définir f(x,y) = 2x² + 3y² + 2xy - 4y
f = 2*x**2 + 3*y**2 + 2*x*y - 4*y

# définir g(x,y) = x³ - 3xy²
g = x**3 - 3*x*y**2

# définir h(x,y) = exp(x³/3 + x - y²)
h = sp.exp(x**3/3 + x - y**2)

# ==========================
# 📌 Affichage des fonctions
# ==========================
# afficher f en LaTeX
display(Math(r"f(x,y) = " + sp.latex(f)))

display(Math(r"g(x,y) = " + sp.latex(g)))

display(Math(r"h(x,y) = " + sp.latex(h)))
```

## 2. Calcul des dérivées partielles (gradient)
Le gradient est :
$\nabla f = (\frac{\partial f}{\partial x}, \frac{\partial f}{\partial y})$.

```python
# ==========================
# 📌 Dérivées partielles
# ==========================
# calculer les dérivées partielles de f
df_dx = sp.diff(f, x)
df_dy = sp.diff(f, y)

# calculer les dérivées partielles de g
dg_dx = sp.diff(g, x)
dg_dy = sp.diff(g, y)

# calculer les dérivées partielles de h
dh_dx = sp.diff(h, x)
dh_dy = sp.diff(h, y)

# ==========================
# 📌 Affichage des gradients
# ==========================
print("📌 Les dérivées partielles des trois fonctions sont :\n")

display(Math(r"\nabla f(x,y) = (" 
             + sp.latex(df_dx) + ", " 
             + sp.latex(df_dy) + ")"))

display(Math(r"\nabla g(x,y) = (" 
             + sp.latex(dg_dx) + ", " 
             + sp.latex(dg_dy) + ")"))

display(Math(r"\nabla h(x,y) = (" 
             + sp.latex(dh_dx) + ", " 
             + sp.latex(dh_dy) + ")"))
```

## 3. Tracé des surfaces des fonctions

```python
# ==========================
#  Surface 3D d’une fonction
# ==========================

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# ==========================
# 📌 Domaine
# ==========================
# générer les valeurs de x entre -3 et 3 (100 points)
X = np.linspace(-3, 3, 100)

# générer les valeurs de y entre -3 et 3 (100 points)
Y = np.linspace(-3, 3, 100)

# créer la grille 2D (meshgrid)
Xg, Yg = np.meshgrid(X, Y)

# ==========================
# 📌 Définition des fonctions
# ==========================
# définir F(x,y) = 2x² + 3y² + 2xy - 4y
F = 2*Xg**2 + 3*Yg**2 + 2*Xg*Yg - 4*Yg

# (optionnel) G(x,y)
G = Xg**3 - 3*Xg*Yg**2

# (optionnel) H(x,y)
H = np.exp(Xg**3/3 + Xg - Yg**2)

# ==========================
# 📌 Création de la figure 3D
# ==========================
# créer une figure de taille (10,7)
fig = plt.figure(figsize=(10, 7))

# ajouter un axe 3D
ax = fig.add_subplot(111, projection='3d')

# ==========================
# 📌 Tracé de la surface
# ==========================
# tracer la surface de F avec transparence alpha=0.6
ax.plot_surface(Xg, Yg, F, alpha=0.6, cmap='viridis')

# ==========================
# 📌 Personnalisation
# ==========================
# ajouter un titre
ax.set_title("Surface de F(x,y)")

# nommer l’axe x
ax.set_xlabel("x")

# nommer l’axe y
ax.set_ylabel("y")

# ==========================
# 📌 Affichage
# ==========================
plt.show()
```

## 4. Affichage d’un point particulier (0,0)

```python
# ==========================
# 📌 Figure 3D
# ==========================
# créer la figure (10,7)
fig = plt.figure(figsize=(10, 7))

# ajouter un axe 3D
ax = fig.add_subplot(111, projection='3d')

# ==========================
# 📌 Tracé de la surface
# ==========================
# tracer la surface F avec alpha=0.5
ax.plot_surface(Xg, Yg, F, alpha=0.5, cmap='viridis')

# ==========================
# 📌 Point particulier
# ==========================
# calculer f(0,0) avec SymPy
z0 = float(f.subs({x: 0, y: 0}))

# afficher le point (0,0,f(0,0)) en rouge
ax.scatter(0, 0, z0, color='red', s=50)

# ==========================
# 📌 Mise en forme
# ==========================
ax.set_title("Point (0,0) sur la surface de f")
ax.set_xlabel("x")
ax.set_ylabel("y")

# ==========================
# 📌 Affichage
# ==========================
plt.show()
```
