## 1. Définition des fonctions  

## Soit les fonctions réelles :

\[
\begin{aligned}
f(x) &= 2x^2 - 3, \\
g(x) &= x^3 - 3x^2 + 2, \\
h(x) &= x^5 + 4x^3 + 4x.
\end{aligned}
\]

```python
import sympy as sp
from IPython.display import display, Math

# ==========================
# 📌 Définir la variable
# ==========================
# définir x comme variable réelle
x = sp.symbols('x', real=True)

# ==========================
# 📌 Définir les fonctions
# ==========================

# f(x) = 2x^2 - 3
f = 2*x**2 - 3

# g(x) = x^3 - 3x^2 + 2
g = x**3 - 3*x**2 + 2

# h(x) = x^5 + 4x^3 + 4x
h = x**5 + 4*x**3 + 4*x

# ==========================
# 📌 Affichage des résultats
# ==========================

print("📌 Les trois fonctions sont :\n")

display(Math(r"f(x) = " + sp.latex(f)))
display(Math(r"g(x) = " + sp.latex(g)))
display(Math(r"h(x) = " + sp.latex(h)))
```

## 2. Calcul de la dérivée première de f

```python
# ==========================
# 📌 Calcul de la dérivée
# ==========================
# calculer la dérivée de f par rapport à x
df = sp.diff(f, x)

# ==========================
# 📌 Affichage du résultat
# ==========================
print("📌 Dérivée de f(x) :")

display(Math(r"f'(x) = " + sp.latex(df)))
```

## 3. Développement limité de f à l’ordre 4 autour de 0

```python
# ==========================
# 📌 Développement limité
# ==========================
# calculer le développement limité de f en 0 à l'ordre 3
DL_f = sp.series(f, x, 0, 4).removeO()

# ==========================
# 📌 Affichage du résultat
# ==========================
print("📌 Développement limité de f(x) en 0 à l'ordre 3 :")
display(Math(r"T_3(x) = " + sp.latex(DL_f)))
```

## 4. Tracé des courbes Cf et Cg avec Matplotlib

```python
# ==========================
# Tracé de fonctions avec Matplotlib
# ==========================

import numpy as np
import matplotlib.pyplot as plt

# ==========================
# 📌 Génération des données
# ==========================
# générer un vecteur X de -3 à 3 avec 400 points
X = np.linspace(-3, 3, 400)

# ==========================
# 📌 Définition des fonctions
# ==========================
# définir Yf = f(x) = 2x² - 3
Yf = 2*X**2 - 3

# définir Yg = g(x) = x³ - 3x² + 2
Yg = X**3 - 3*X**2 + 2

# ==========================
# 📌 Tracé des courbes
# ==========================
# tracer f(x)
plt.plot(X, Yf, label='f(x) = 2x^2 - 3')

# tracer g(x)
plt.plot(X, Yg, label='g(x) = x^3 - 3x^2 + 2')

# ==========================
# 📌 Personnalisation du graphe
# ==========================
plt.xlabel('x')
plt.ylabel('y')
plt.title('Courbes de f(x) et g(x)')

# ==========================
# 📌 Finalisation
# ==========================
plt.legend()
plt.grid(True)
plt.show()
```

## 5. Affichage des points (0,g(0)) et (2,g(2))

```python
import numpy as np
import matplotlib.pyplot as plt

# définir la fonction g(x)
x = sp.symbols('x', real=True)
g = x**3 - 3*x**2 + 2

# ==========================
# 📌 Domaine de tracé
# ==========================
# générer X entre -3 et 3 (400 points)
X = np.linspace(-3, 3, 400)

# calculer Yg (version numérique de g(x))
g_num = sp.lambdify(x, g, 'numpy')
Yg = g_num(X)

# ==========================
# 📌 Tracé de la courbe
# ==========================
plt.plot(X, Yg, label='g(x)')

# ==========================
# 📌 Points particuliers
# ==========================
# calculer g(0) et g(2) avec SymPy (subs + float si besoin)
y0 = float(g.subs(x, 0))
y2 = float(g.subs(x, 2))

# afficher les points (0, g(0)) et (2, g(2))
plt.scatter([0, 2], [y0, y2], color='red')

# ==========================
# 📌 Annotation des points
# ==========================
plt.text(0, y0, '(0, g(0))')
plt.text(2, y2, '(2, g(2))')

# ==========================
# 📌 Mise en forme du graphique
# ==========================
plt.xlabel('x')
plt.ylabel('y')
plt.title('Points particuliers sur g(x)')
plt.grid(True)
plt.legend()
plt.show()
```

## 7. Tracé de h'(x)

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# ==========================
# 📌 Variable symbolique et fonction
# ==========================
# définir x comme variable réelle SymPy
x = sp.symbols('x', real=True)

# définir la fonction h(x)
h = x**5 + 4*x**3 + 4*x

# ==========================
# 📌 Dérivée
# ==========================
# calculer la dérivée dh = h'(x)
dh = sp.diff(h, x)

# ==========================
# 📌 Domaine numérique
# ==========================
# générer X entre -3 et 3 (400 points)
X = np.linspace(-3, 3, 400)

# ==========================
# 📌 Évaluation numérique de la dérivée
# ==========================
# calculer Ydh = dh(x) pour chaque valeur de X
Ydh = sp.lambdify(x, dh, 'numpy')(X)

# ==========================
# 📌 Tracé
# ==========================
plt.plot(X, Ydh, label="h'(x)")

# ==========================
# 📌 Mise en forme
# ==========================
plt.xlabel('x')
plt.ylabel("h'(x)")
plt.title("Dérivée de h(x)")
plt.grid(True)

plt.legend()

# afficher le graphique
plt.show()
```
